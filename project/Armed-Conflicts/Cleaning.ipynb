```python
import pandas as pd
import numpy as np
import seaborn as sns 

import matplotlib.pyplot as plt
```

```python
conflicts_raw = pd.read_csv('conflicts.csv')  
#conflicts['year'] =  pd.to_datetime(conflicts['year'],format= '%Y')
conflicts_raw.head()
```

```python
conflicts = conflicts_raw.drop(['location','side_a_2nd','side_b_2nd','side_a_id','side_b_id','territory_name',
                                'cumulative_intensity','ep_end_prec','start_prec','start_date2','start_prec2','gwno_a','gwno_loc',
                                'gwno_a_2nd','gwno_b','gwno_b_2nd','region','version'], axis=1)


conflicts= conflicts.sort_values(by=['start_date'])
conflicts.head()
```

```python

#pd.set_option('display.max_rows', None)


#for loop for incompatibility, intensity, type, region
incompatible = [] 
for value in conflicts["incompatibility"]: 
    if value == 1: 
        incompatible.append("Territory") 
    elif value == 2: 
        incompatible.append("Government") 
    else: 
        incompatible.append("Government and Territory")
        
conflicts['incompatibility'] = incompatible

#intensity
intense = [] 
for value in conflicts["intensity_level"]: 
    if value == 1: 
        intense.append("Minor")  
    else: 
        intense.append("War")
        
conflicts['intensity_level'] = intense

#type of conflict
types = [] 
for value in conflicts["type_of_conflict"]: 
    if value == 1: 
        types.append("Extrasystemic") 
    elif value == 2: 
        types.append("Interstate") 
    elif value ==3:
        types.append('Internal')
    else: 
        types.append("Internationalized Internal")
        
conflicts['type_of_conflict'] = types
```

```python
conflicts.set_index('conflict_id')
```

```python
#Clean sides
def clean_country(country):
    return (country
            .replace('Government of ', '')
            .replace(' (Soviet Union)',''))

conflicts['side_a'] = [clean_country(country) for country in conflicts['side_a']]
conflicts['side_b'] = [clean_country(country) for country in conflicts['side_b']]

#conflicts = conflicts.groupby(['side_a','conflict_id']).first()
#conflicts.head()
```

```python
gdp_raw = pd.read_csv('country_gdp.csv')
```

```python
pop_raw = pd.read_csv('population.csv')
```

```python
literacy_raw = pd.read_csv('literacy.csv')
literacy_raw['last_value'] = literacy_raw.ffill(axis=1).iloc[:, -1] 
```

```python
military_raw = pd.read_csv('military.csv')
```

```python
gini_raw = pd.read_csv('gini.csv')
gini_raw['last_value'] = gini_raw.ffill(axis=1).iloc[:, -1] 
```

```python
cdr_raw = pd.read_csv('cdr.csv')
```

```python
life_expectancy_raw = pd.read_csv('life_exp.csv')
```

```python
refugee_raw = pd.read_csv('refugee.csv')
```

```python
child_mortality_raw = pd.read_csv('child_mortality.csv')
```

```python
pop_drop = pop_raw[['Country Name']].rename(columns={'Country Name':'side_a'})
```

```python
#conflicts = conflicts.merge(indicators,on='side_a')
#conflicts = conflicts.groupby(['side_a','conflict_id']).first()

conflicts['year'] = conflicts['year'].apply(str)
conflicts = conflicts.merge(pop_drop,on='side_a')
conflicts
```

```python
conflicts_1960 = conflicts.loc[conflicts['year'] > '1959']
```

```python
pops = []
gdp = []
militia = []
gini = []
lit = []
cdr = []
life = []
child = []
refuge = []

conflicts_grouped = conflicts_1960['side_a'].unique()
for country in conflicts_grouped:
    table = conflicts_1960.loc[conflicts_1960['side_a'] == country]
    for year in table['year']:
        pops.append(pop_raw.loc[pop_raw['Country Name'] == country, year].values[0])
        gdp.append(gdp_raw.loc[gdp_raw['Country Name'] == country, year].values[0])
        militia.append(military_raw.loc[military_raw['Country Name'] == country, year].values[0])
        gini.append(gini_raw.loc[gini_raw['Country Name'] == country, year].values[0])
        lit.append(literacy_raw.loc[literacy_raw['Country Name'] == country, year].values[0])
        cdr.append(cdr_raw.loc[cdr_raw['Country Name'] == country, year].values[0])
        life.append(life_expectancy_raw.loc[life_expectancy_raw['Country Name'] == country, year].values[0])
        child.append(child_mortality_raw.loc[child_mortality_raw['Country Name'] == country, year].values[0])
        refuge.append(refugee_raw.loc[refugee_raw['Country Name'] == country, year].values[0])
        

conflicts_1960['GDP'] = gdp
conflicts_1960['Population'] = pops
conflicts_1960['Military Expenditure % GDP'] = militia
conflicts_1960['Gini Index'] = gini
conflicts_1960['Adult Literacy Rate'] = lit
conflicts_1960['Crude Death Rate'] = cdr
conflicts_1960['Child Mortality Rate'] = child
conflicts_1960['Life Expectancy'] = life
conflicts_1960['Refugees from Origin'] = refuge
```
