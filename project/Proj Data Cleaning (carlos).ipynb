```python
data_no_red = data[data['HouseD_2014'] != "      Redrawn"]
data_no_red
```

```python
data_no_red.info()
```

```python
data_no_red['HouseD_2014'] = pd.to_numeric(data_no_red["HouseD_2014"], downcast="float")
data_no_red['HouseD_2012'] = pd.to_numeric(data_no_red["HouseD_2012"], downcast="float")
```

```python
# checking that types are now changed to numerical values
data_no_red.info()
```

```python
# extracting out the numerical columns 
num_data_no_red = data_no_red.select_dtypes(exclude=[object])
num_data_no_red.info()
```

```python
num_data_no_red
```

```python
num_data_no_red.drop(['HouseR_2012', 'HouseR_2014'], axis=1).info()
```

```python
# Building a linear regression model to try to predict HouseR_2014 from all other numeric variables 
# excluding HouseR_2012 which we will predict later on. 


X = num_data_no_red.drop(['HouseR_2012', 'HouseR_2014'], axis=1).iloc[:, 0:12]
Y = num_data_no_red['HouseR_2014']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, explained_variance_score

model = linear_model.LinearRegression()

# Training the model 
model.fit(X_train, y_train)

predicted_y = model.predict(X_test)

model2014 = model

mse = mean_squared_error(predicted_y, y_test)

# Explained variance regression score function
# Best possible score is 1.0, lower values are worse.

variance = explained_variance_score(y_test, predicted_y)


mse, variance
```

```python
# Building a linear regression model to try to predict HouseR_2012 from all other numeric variables 
# excluding HouseR_2014


X = num_data_no_red.drop(['HouseR_2012', 'HouseR_2014'], axis=1).iloc[:, 0:12]
Y = num_data_no_red['HouseR_2012']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)


model = linear_model.LinearRegression()

# Training the model 
model.fit(X_train, y_train)

predicted_y = model.predict(X_test)

model2012 = model

mse = mean_squared_error(predicted_y, y_test)

# Explained variance regression score function
# Best possible score is 1.0, lower values are worse.

variance = explained_variance_score(y_test, predicted_y)


mse, variance
```

```python
# Now we will use our models to predict the HouseR_2014 and HouseR_2012 columns

null = data[data['HouseR_2014'].isnull() == True]

num_data = null.select_dtypes(exclude=[object]).iloc[:, 0:12]

predicted_2014 = model2014.predict(num_data)

predicted_2012 = model2012.predict(num_data)
```

```python
null.loc[:, 'HouseR_2014'] = predicted_2014

null.loc[:, 'HouseR_2012'] = predicted_2012
```
