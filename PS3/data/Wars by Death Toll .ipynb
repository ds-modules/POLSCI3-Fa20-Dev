## Part 4 
Giving each country an estimated high or low GDP value based on date and median GDP of the time

```python
years = gdp_death.iloc[:, 6:15]
years=years.dropna(0)

def no_nulls(x):
    if (x !='-'):
        return x
    pass
med1850 = np.median([int(x) for x in [no_nulls(x) for x in years['1850']] if x is not None])
med1870 = np.median([int(x) for x in [no_nulls(x) for x in years['1870']] if x is not None])
med1913 = np.median([int(x) for x in [no_nulls(x) for x in years['1913']] if x is not None])
med1929 = np.median([int(x) for x in [no_nulls(x) for x in years['1929']] if x is not None])
med1937 = np.median([int(x) for x in [no_nulls(x) for x in years['1937']] if x is not None])
med1960 = np.median([int(x) for x in [no_nulls(x) for x in years['1960']] if x is not None])
med1973 = np.median([int(x) for x in [no_nulls(x) for x in years['1973']] if x is not None])
med1995 = np.median([int(x) for x in [no_nulls(x) for x in years['1995']] if x is not None])
med2016 = np.median([int(x) for x in [no_nulls(x) for x in years['2016']] if x is not None])
```

```python
ind = 0
high_low = []
for x in gdp_d['mid']:

    if (gdp_d.iloc[ind,6] == 'low'):
        high_low.append('low')

    else:
        if (x > 1800 and x <= 1860):
            if (((med1850 + med1870)/2) <= int(gdp_d.iloc[ind,6])):
                high_low.append('high')
            else:
                high_low.append('low')
        elif (x > 1860 and x <= 1891.5):
            if (((med1870 + med1913)/2) <= int(gdp_d.iloc[ind,6])):
                  high_low.append('high')
            else:
                high_low.append('low')
        elif (x > 1891.5 and x <= 1921):
            if (((med1913 + med1929)/2)  <= int(gdp_d.iloc[ind,6])):
                high_low.append('high')
            else:
                high_low.append('low')
        elif (x > 1921 and x <= 1933):
            if (((med1929 + med1937)/2) <= int(gdp_d.iloc[ind,6])):
                high_low.append('high')
            else:
                high_low.append('low')
        elif (x > 1933 and x <= 1948.5):
            if (((med1937 + med1960)/2) <= int(gdp_d.iloc[ind,6])): 
                high_low.append('high')
            else:
                high_low.append('low')
        elif (x> 1948.5 and x<= 1966.5):
            if (((med1960 + med1973)/2) <= int(gdp_d.iloc[ind,6])):
                high_low.append('high')
            else:
                high_low.append('low')
        elif (x> 1966.5 and x<= 1984):
            if (((med1973 + med1995)/2) <= int(gdp_d.iloc[ind,6])):
                high_low.append('high')
            else:
                high_low.append('low')
        elif (x>1984 and x<= 2005.5):
            if (((med1995 + med2016)/2) <= int(gdp_d.iloc[ind,6])):
                high_low.append('high')
            else:
                high_low.append('low')
        elif (x > 2005.5):
            if med2016 >= x:
                high_low.append('high')
            else:
                high_low.append('low')
    ind += 1
gdp_d['high_low'] = high_low        
```
